##DAY-5

**Write a program to count the frequency of each word in a text. Print the word with the corresponding count**

```python

text = input("Enter a text: ")


words = text.lower().split()


word_count = {}


for word in words:
    word = word.strip(",.?!")
    word_count[word] = word_count.get(word, 0) + 1


print("\nWord Frequency:")
for word, count in word_count.items():
    print(f"{word}: {count}")
```

##DAY-6

**Write a Python program to using NLTK and Spacy
Convert text to lowercase.
Remove stopwords using NLTK**


```python
import nltk
from nltk.corpus import stopwords
import spacy

nltk.download('stopwords')


nlp = spacy.load("en_core_web_sm")


text = "This is an example sentence to demonstrate removing stopwords using NLTK and spaCy."


doc = nlp(text.lower())
print("Lowercased Text:")
print(doc.text)


stop_words = set(stopwords.words("english"))
filtered_words = [token.text for token in doc if token.text not in stop_words]

print("\nText after Removing Stopwords:")
print(" ".join(filtered_words))
```

**Write a Python script that: 1. Use Genism to preprocess data from a sample text file, follow basic procedures like tokenization, stemming, lemmatization etc**

```python

!pip install gensim nltk spacy

import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
!python -m spacy download en_core_web_sm

import gensim
from gensim.utils import simple_preprocess
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
import spacy
from google.colab import files

nlp = spacy.load('en_core_web_sm')

stemmer = SnowballStemmer("english")
stop_words = set(stopwords.words("english"))


def preprocess_text(text):
    """
    Preprocess text data.

    Steps include:
    - Tokenization
    - Removing stop words
    - Stemming
    - Lemmatization
    """
    tokens = simple_preprocess(text, deacc=True)


    tokens = [word for word in tokens if word not in stop_words]


    stemmed_tokens = [stemmer.stem(word) for word in tokens]


    doc = nlp(" ".join(tokens))
    lemmatized_tokens = [token.lemma_ for token in doc if token.lemma_ not in stop_words]


    print("\nOriginal Text:\n", text[:200], "...")
    print("\nTokenized:\n", tokens[:20])
    print("\nStemmed Tokens:\n", stemmed_tokens[:20])
    print("\nLemmatized Tokens:\n", lemmatized_tokens[:20])

    return lemmatized_tokens


print("Please upload a text file:")
uploaded = files.upload()

for file_name in uploaded.keys():
    with open(file_name, 'r', encoding='utf-8') as file:
        text_data = file.read()


preprocessed_data = preprocess_text(text_data)

print("\nFinal Preprocessed Data:\n", preprocessed_data)
```

#DAY-8
**Write a Python script that:Tokenizes a sample paragraph into words and sentences.**

```python
# Install necessary libraries (for Colab or locally)
!pip install nltk spacy

# Download necessary NLTK resources
import nltk
nltk.download('all')

# Import libraries
from nltk.tokenize import word_tokenize, sent_tokenize

# Sample paragraph for tokenization
sample_paragraph = """
Artificial Intelligence (AI) is transforming the world.
From healthcare to education, AI-powered solutions are making processes efficient and improving decision-making.
It's a fascinating era for technology enthusiasts!
"""

# Tokenize sentences using NLTK
sentences = sent_tokenize(sample_paragraph)
print("Tokenized Sentences:")
print(sentences)

# Tokenize words using NLTK
words = word_tokenize(sample_paragraph)
print("\nTokenized Words:")
print(words)

# Optional: Use spaCy for tokenization (if installed and preferred)
import spacy

# Load spaCy's English model
nlp = spacy.load('en_core_web_sm')

def spacy_tokenization(text):
    # Process the text with spaCy
    doc = nlp(text)

    # Sentence tokenization
    spacy_sentences = [sent.text for sent in doc.sents]
    print("\nTokenized Sentences (spaCy):")
    print(spacy_sentences)

    # Word tokenization
    spacy_words = [token.text for token in doc]
    print("\nTokenized Words (spaCy):")
    print(spacy_words)

# Tokenize using spaCy
spacy_tokenization(sample_paragraph)
```
